```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Time_series_analysis_and_forecast_DATASET.csv')
```

```python
data.head()
```

```python
data['FullDate'] = pd.to_datetime(data['FullDate'])
```

```python
data.info()
```

```python
data.set_index("FullDate", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(data.index, data.ElecPrice, '--', marker='*', )
plt.grid()
plt.xlabel('FullDate')
plt.ylabel('ElecPrice')
```

```python
data_daily = data.resample('D').mean()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data_daily.index, data_daily.ElecPrice, '--', marker='*', )
plt.grid()
plt.xlabel('FullDate')
plt.ylabel('ElecPrice')
```
